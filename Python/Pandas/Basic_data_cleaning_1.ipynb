**Pandas :** Pandas is a python libray for data manipulation and analysis.

> Thinks of it as **"Excel for Python"** - which helps to load, clean, filter, sort and analyze data using tables called DataFrames. 

**Why pandas?**
* Fast - handles millions of rows easily.
* Simple - readable, clean syntax.
* Flexible - works with CSV, Excel, JSON, SQL and  More.
* Powerful - built-in tools for cleaning, grouping, merging data.
* Popular - used in data science, ML, finance, research everywhere.
* Built on NumPy - so, it's numerically every efficient.

**📚 Pandas Full Learning Path :**

1. Series          → 1D data (like a single column)

```python
Score = pd.Series([85, 90, 78, 92, 88, 70, 95, 80, 73, 91])
print(Score)
```

**2. DataFrame (Two-Dimensional Data) :** Full Excel (Column & Row)

**Required another column for two-dimensional data:**

```python
Name = pd.Series(["Ali", "Bob", "Sara", "John", "Mia",  "Tom", "Emma", "Ray", "Nina", "Leo"])
print(Name)
```

**One-way of creating DataFrame :**

```python
df = pd.DataFrame({
      "Name" : Name,
      "Score" : Score
})
print(df)
```

```python
df = pd.DataFrame({
          "Age" : ["M", "M", "F", "M", "F", "M", "F", "M", "F", "M"],
          "Department" : ["CSE", "BBA", "CSE", "EEE", "BBA", "CSE", "EEE", "BBA", "EEE", "CSE"]  
})
df
```

**5. Selecting Data :**  columns, rows, loc, iloc, at, iat, set_index
          
          * single column               ->  ["col"]	
          * Multiple columns            ->  ["col1", "col2"]         
          * Select by label/name        ->  .loc[]
          * select by position number   ->  .iloc[]

          
          * label	single fast value     ->  .at[]
          * position	single fast value ->  .iat[]
          * setting index               ->  .set_index

**a. Selecting single column:**

```python
df['Name']
```

**b. Selecting multiple column:**

```python
df[['Name', 'Score']] 
```

**c. Selecting Rows :** 

position-based slicing, end excluded → only row 0

```python
df[:1]   
```

```python
df[0:5]
```

**d. Label-based selection :**  Syntax -> df.loc[row_label, column_label]

**Single row with specific column:**

label-based slicing, end included → row 0 and 1

```python
df.loc[:1]
```

**Specific columns:**

```python
df.loc[0:2, ['Name', 'Score']]
```

**e. Position-based selection:** Use index position.

 Syntex-> df.iloc[row_index, column_index]

**single row & column:**

```python
df.iloc[0,1]
```

**Multiple rows & columns:**

```python
df.iloc[0:2, 0:2]
```

**f. Selecting Specific Values:**

With specific label:

```python
df.at[0, 'Score']
```

With specific position :

```python
df.iat[0,1]
```

**6. Filtering →** showing only rows that match a condition.

**a. Boolean Selection:**


```python
df[df['Score'] > 80]
```

**b. Selecting With conditions and loc:**

```python
df.loc[df['Score'] > 80, ['Name']]
```

**c. More filters by using conditions:**

**Not equal:**

```python
df[df["City"] != "Dhaka"]  
```

**Opposite:**

```python
df[~(df['Gender'] == "M")]
```

```python
df[df["Score"].between(87, 90)]
```

**isin() :**

```python
df[df["City"].isin(["Dhaka", "London"])]
```

**c. Query Method:** Advanced

```python
df.query("Score > 80 and Department == 'EEE'")
```
